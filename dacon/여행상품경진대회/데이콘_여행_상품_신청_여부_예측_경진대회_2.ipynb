##2. 구현 방법 선정

여행 상품 신청 여부 예측이므로 각 요소에 얼마의 사람들이 얼마나 상품을 신청할지 안할지 구해야 한다.

그럼 모델링을 하기 앞서 우리는 어떤 모델을 구현해야 할까?

우선 우리는 상품 신청을 **예측**해야 한다. 

따라서 머신러닝 기술중 연속적인 숫자를 예측하는 회귀 분석을 사용해야 한다.


* tip 회귀 분석이란 : 규명된 함수식을 이용해서 종속변수의 변화를 예측 하는 분석

###회귀 분석에 대한 종류

회귀 분석은 크게 5가지로 나누어 진다. (대표적으로 유명한 것)
+ 선형 회귀(Linear regression) ...
+ 로지스틱 회귀(Logistic regression) ...
+ 리지 회귀(Ridge regression) ...
+ 라쏘 회귀(Lasso regression) ...
+ 다항 회귀(Polynomial regression)

어떤 방법을 써야 모델을 잘만들었다고 소문이 날까?

음... 데이터를 살펴보니 눈에 확 보이는 데이터의 특징이 있습니다. 혹시 찾으셨나요?

자 바로 데이터의 속성이 섞여 있다는 것을 알 수 있습니다.

잘 모르시는 분들을 위해서 설명하자면 데이터는 여러가지 기준으로 나눌 수 있지만 데이터의 속성에 따라 범주형데이터와 연속형 데이터로 나눌 수 있습니다.

그럼 범주형 데이터란 뭘까?
범주형 데이터란 **범주를 서로 구분하는 이름이 있는 데이터**를 의미힙나다.
예를 들어 남/여, 전문직/비전문직,아버지/어머니로 나누어 지는 걸 말합니다.

이제 반대로 연속형 데이터는 **연속적인 수로 수량화가 가능한 데이터**라고 설명할 수 있습니다.
예를 들어 키, 온도, 나이 등 수량화가 가능한 것들을 말합니다.

그럼 이제 본론으로 들어가서 지금 주어진 데이터들을 살펴보면 범주형과 연속형 데이터가 섞여 있지만 대부분 여행 패키지 신청 여부(0,1), 주거 중인 도시의 등급(1,2,3),영업 사원이 제시한 상품등 서로 구분이 가능한 범주형데이터가 주를 이루고 있습니다.

물론 나이, pt시간등 연속형 데이터가 존재하기는 하지만 많은 수의 데이터들이 범주형 데이터를 가지고 있는 것을 보아서 우선 우리는 범주형 데이터에 적합한 모델을 만들어보고 만약 부족하다면 연속형 데이터에 적합한 모델을 만들어 보면 좋을 것 같습니다.

그런 이유로 저희는 우선 범주형 데이터에 적합한 로지스틱 회귀모델을 사용해야 합니다.
